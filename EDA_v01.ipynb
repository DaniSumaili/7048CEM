```python
# check for label distribution
label_count = breastdata['type'].value_counts()
label_count
```

```python
x= breastdata['type'].value_counts().index.tolist()
y= label_count.values

plt.title('Cancer Sub-types distribution')
plt.bar(x,y, color=["blue", "orange","green", "red", "purple", "brown"])
for i in range(len(x)):
  plt.text(i,y[i],y[i], ha="center",va="bottom")

plt.show()
```
